```python
import pandas as pd
def merge_csv_files(file_paths, output_file):
    dataframes = []
    for i, file_path in enumerate(file_paths):
        df = pd.read_csv(file_path)
        df['source'] = file_path
        if i == 0:
            column_headers = list(df.columns)
        else:
            assert column_headers == list(df.columns)
        df.columns = [col for col in df.columns]
        dataframes.append(df)
    merged_df = pd.concat(dataframes, axis=0)
    # merged_df.columns = [col.split("_", 1)[1] for col in merged_df.columns]
    merged_df.to_csv(output_file, index=False, columns=column_headers)

def findCSV_inFolder():
    import os
    import glob
    csv_files = []
    os.chdir(".") # 切换到当前目录，如果不需要可以省略该行代码
    for file in glob.glob("*.csv"):
        csv_files.append(file)
    csv_files.sort()
    print(csv_files)
    return csv_files
## 1
# prefix = ''
# ls = [prefix+str(i)+'.csv' for i in range(0, 8)]
# file_paths = ['1.csv', '2.csv', '3.csv', '4.csv']
## 2
ls = findCSV_inFolder()
output_file = 'best.csv'
merge_csv_files(ls, output_file)
# print(ls)
```
